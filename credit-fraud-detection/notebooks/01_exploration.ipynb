# 💳 Détection de défaut de paiement — Exploration

Analyse exploratoire du dataset de transactions bancaires.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

COLUMNS = ['Time'] + [f'V{i}' for i in range(1, 29)] + ['Amount', 'Class']
df = pd.read_csv('../data/creditcard.csv', header=None, skiprows=1, names=COLUMNS)
df.dropna(subset=['Class'], inplace=True)
df['Class'] = df['Class'].astype(int)
print(f'Shape : {df.shape}')
df.head()
```

## 1. Distribution des classes

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

counts = df['Class'].value_counts()
axes[0].bar(['Normal (0)', 'Fraude (1)'], counts.values, color=['#0F6E56', '#D85A30'])
axes[0].set_title('Distribution des classes')
axes[0].set_ylabel('Nombre de transactions')
for i, v in enumerate(counts.values):
    axes[0].text(i, v + 5, str(v), ha='center', fontweight='bold')

axes[1].pie(counts.values, labels=['Normal', 'Fraude'],
            colors=['#0F6E56','#D85A30'], autopct='%1.2f%%', startangle=90)
axes[1].set_title('Proportion fraudes / normales')

plt.tight_layout()
plt.show()
```

## 2. Distribution des montants

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

for label, color, ax in zip([0, 1], ['#0F6E56', '#D85A30'], axes):
    subset = df[df['Class'] == label]['Amount']
    ax.hist(subset, bins=40, color=color, alpha=0.85, edgecolor='white')
    ax.set_title(f"Montant — {'Normal' if label==0 else 'Fraude'}")
    ax.set_xlabel('Montant (€)')
    ax.set_ylabel('Fréquence')
    ax.axvline(subset.mean(), color='black', linestyle='--', label=f'Moyenne : {subset.mean():.2f}€')
    ax.legend()

plt.tight_layout()
plt.show()
```

## 3. Statistiques descriptives

```python
print('=== Statistiques générales ===')
print(df[['Amount', 'Time']].describe().round(2))
print(f"\nTaux de fraude : {df['Class'].mean()*100:.2f}%")
```

## 4. Corrélation des features avec la classe

```python
corr = df.corr()['Class'].drop('Class').sort_values()
fig, ax = plt.subplots(figsize=(14, 5))
colors = ['#D85A30' if v < 0 else '#0F6E56' for v in corr.values]
ax.bar(corr.index, corr.values, color=colors)
ax.axhline(0, color='black', linewidth=0.8)
ax.set_title('Corrélation de chaque feature avec la variable cible (Class)')
ax.set_xticklabels(corr.index, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
